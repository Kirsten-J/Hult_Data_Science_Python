```python
# import libraries
import random
from PIL import Image
import urllib.request
from IPython.core.display import display, HTML
import sys

# start game function. I did not count this as a stage
## I have text wrapping in my cells, so the text will likely look like a long line for others, but for me it is automatically wrapped to fit the screen.
def game():
    """This function asks if you want to play and starts the game. The game is based on toy story 4. \nIn it, you are Woody and must save Forky from Gabby Gabby's clutches. Most decisions will result \nin a win, although a few will cause you to be caught and therefore fail the mission. The shortest way to \nfinish this game takes under 3 minutes, but some decisions will make you take more than 3 minutes to \ncomplete it. Some inputs are printed before the text even though it is ordered correctly in the function. \nAlso, in some parts, if a wrong input is given it will return to the last checkpoint, \nnot just that question because it starts the stage all over again. \nI did not count this as a stage.\n  """
    
    # https://stackoverflow.com/questions/25698448/how-to-embed-html-into-ipython-output display html
    display(HTML(data = """<h1>Text Adventure Game: Forky's Escape </h1> <p>Welcome to my text adventure game. This game is based on a scene from Toy Story 4 where the toys must save Forky from evil Gabby Gabby. I hope you enjoy playing it. If at any point you give a wrong input it will return to the last checkpoint, but if you lose you will have to start from the beginning. </p> <html>"""))
    print(f"""You are Woody, a cowboy toy who has recently been given to charity by your owner, Andy. \nYou made a lot of friends in your new home, but one of them, Forky, has been kidnapped \nby Gabby Gabby's henchman Benson, a dummy. \n""")
    
    #ask if they want to play exit = leave, else start game
    in1 = input(prompt=f"""Press any key to begin, or type "exit" to end. Keep in mind sometimes the \ninput area will be printed above the game text, so just scroll up to see it. Each part \nshould be separated by a line of ">" so you can know from where to read easily. \n{">"*80} """)
    if in1 == "exit":
        return print(f"""See you later!""")
    else: #starting paragraph and image
        hostage1 = "https://www.indiewire.com/wp-content/uploads/2019/06/Screen-Shot-2019-06-13-at-10.36.43-AM.png"
        hostage2 = "https://i.ytimg.com/vi/ZBAjq_dLQSM/hqdefault.jpg"
        hostage3 = "https://pbs.twimg.com/media/D7YOYl3XkAAoZnB.jpg"
        hostages = [hostage1, hostage2, hostage3]
        image(random.choice(seq = hostages))
        print("""\nYou reach the store where Gabby Gabby is keeping him hostage, hoping to save him from her clutches. \nBuzz Lightyear, Bo Peep, Bunny and Ducky are with you and may help you in your mission. \n""")
        print(f"""Once inside, you stare at three long aisles looming in front of you. \nWhich one will you walk through?

    1) Features old furniture. These may be a good way to climb to the top of shelves, \nbut because they are old, they might make noise when climbed.

    2) Old posters and paintings cover the sides of the wall. Some with protective broken \nglass covers could potentially cut you. 

    3) Lamps. Lamps everywhere. They are very fragile so you would have to make sure not to \nknock anything over. \n"""
    )
        print("""\n --- CHECKPOINT --- """)
        arrived(i = 1) #enter stage 1
        

# I did not count this as a stage
# fail function for losing the game
def fail():
    """This function will be called everytime the player fails the game. It asks if they want to play again.""" 
    #image of them caught
    caught1 = "https://static.wikia.nocookie.net/disney/images/a/a5/Bensons.png/revision/latest?cb=20191210003159"
    caught2 = "https://i.ytimg.com/vi/8JDDynfP7Zk/maxresdefault.jpg"
    caught3 = "https://1.bp.blogspot.com/-rffeYzuA5U0/XQ2cVLXc1RI/AAAAAAAAG54/1tpttgMG56k7UAQkKV4ZeIoQUVyFfxAFQCLcBGAs/s1600/benson.jpg"
    caught = [caught1, caught2]
    image(random.choice(seq = caught))
    print(f"""{"*"*80}\n """)
    display(HTML(data = """<h1>YOU LOST</h1> <h2>Do you want to play again?</h2> """))
    play = str(input(prompt=f"""{">"*80} """))
    #check if the input contains any word in each list (repeats all over the game)
    substring = ["yes", "play", "again", "yeah", "yep", "of course", "affirmative"]
    no = ["no", "nope", "nah", "dont", "don't", "negative"]
    notagain = any(word in play.casefold() for word in no)
    again = any(word in play.casefold() for word in substring) # https://www.kite.com/python/answers/how-to-check-if-a-list-contains-a-substring-in-python
    if again == True or play == "y": #if they put play again start the game again
        game()
    elif notagain == True or play == "n": #if not stop running the code
        return print(f"""See you later! \n""")
        
    else: 
        print("""That is not acceptable answer, please try again \n""")
        fail()


# I did not count this as a stage
# function to show an image inline from a url
# https://stackoverflow.com/questions/40911170/python-how-to-read-an-image-from-a-url
def image(fileurl):
    """This function takes a url and saves it as an image to show within jupyter. The user never \ninteracts with this function, it was used to insert the images seen in the game without having to \nwrite so many lines each time."""
    try:
        urllib.request.urlretrieve(url = fileurl, filename = "file.png")  #ask for url
        img = Image.open(fp = "file.png")  #open image from url
        img = img.resize(size = (400, 300))
        display(img)  #show image inline
    except IOError:  #if no working url given
        pass


# I did not count this as a stage
# function for when people win
def win():
    """This function will be called whenever the player wins the game. It will ask if they want to play again."""
    # win message and image
    display(HTML(data = """<h2>Congratulations, you were able to leave the shop and save your friends!</h2>"""))
    win1 = "https://i.pinimg.com/originals/d5/21/6e/d5216e671d6e3a99cd136d01ee83d8b1.png"
    win2 = "https://d23.com/app/uploads/2019/05/1180w-600h_052819_TS4-Meet_Forky-780x440.jpg"
    wins = [win1, win2]
    image(random.choice(seq = wins))

    # ask to play again
    print(f"""{"*"*80} \nDo you want to play again? \n""")
    play = str(input(prompt=f"""{">"*80} """))
    substring = ["yes", "play", "again", "yep", "yeah", "of course"]
    no = ["no", "nope", "nah", "dont", "don't", "negative"]
    again = any(word in play.casefold() for word in substring)
    notagain = any(word in play.casefold() for word in no)
    if again == True or play == "y": #if yes start game
        game()
    elif notagain == True or play == "n":
        return print("""See you later! \n""") #if no stop running the code

    else: # unexpected answer dont know what to do
        print("""That is not acceptable answer, please try again \n""")
        win()


    
### STAGES ###


# STAGE 1
# function for the when you first arrive at the store
def arrived(i=1):
    """This is the first stage. The default arg helps to increase a counter. I can't define i inside the function because it will reset each time. The user does not need to know this argument exists, and will never interact with it."""
    #input choice of where to go, options and check if they wrote one of the options
    arrive = str(input(prompt=f"""{">"*80} """))
    option1 = ["one", "1st", "furniture", "shelves", "first", "fur"]
    option2 = ["two", "second", "poster", "paintings", "glass", "pos", "pain"]
    option3 = ["three", "3rd", "la", "lamp", "behave"]

    optionone = any(word in arrive.casefold() for word in option1)
    optiontwo = any(word in arrive.casefold() for word in option2)
    optionthree = any(word in arrive.casefold() for word in option3)

    #if right choice continue story
    if optionone == True or arrive == "1":
        print(f"""You and your friends run towards the shortest chair and climb until you can see \nthe entire store. Near the rear you notice Gabby Gabby walking out of a cabinet \nfull of antique plates. She is followed by at least three Benson dummies, \nwith one staying behind. This is most likely where Forky is. """)
       # i=[0] # This has to be set as 0 if you chose option 3 the first time but not the second, so it works if you play again
        inside()
    #if wrong choice fail
    elif optiontwo == True or arrive == "2":
        print(f"""You walk down the hall of posters. Some are even in black and white. While admiring one of \nthem you notice Gabby Gabby and Benson looking at you through the reflection in the glass.\n""")
        print(f"""Gabby Gabby and Benson have you in their sights and despite the fact you try to run away you \ncan't leave the store and you get caught. You cannot save Forky as you are also taken hostage.\n""")
       # i=[0] # This has to be set as 0 if you chose option 3 the first time but not the second, so it works if you play again
        fail()

    #third choice also works but makes story longer (also have to press 3 twice for it to send to longer story)
    elif optionthree == True or arrive == "3":
        i = i + 1
        #if chosen twice continue with the story (longer version)
        if i >= 3:
            print("""Bunny and Ducky do not act up this time. You follow the hallway and reach the pantry. \nYou realize Forky is not in there and must continue searching the store. \nYou reach a makeshift maze made of old, tattered books. Gabby Gabby must have thought you would \nenter through the pantry window and set them up. \n """)
            return maze()
        #first time chosen it will ask for the choice again
        else:
            print(f"""You start to walk towards the lamp hallway but notice Bunny and Ducky start to clown around. \n""")
            bunny1 = "https://static.onecms.io/wp-content/uploads/sites/6/2019/06/toy-story-4-ducky-and-bunny-2000.jpg"
            bunny2 = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRayaSgGEPWoIHOpjaqOqV9VgC_ccyQP4GRmA&usqp=CAU"
            bunny3 = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSWFE5pSqlqipsfqn4lIoKjFqtZxYlG2_aGMA&usqp=CAU"
            bunnys = [bunny1, bunny2, bunny3]
            image(random.choice(seq = bunnys))
            print(f"""You can't take that risk, so you make everyone return and give Bunny and Ducky a much-needed talk. \nThey seem to understand the dangers of the situation. Which hallway will you follow now?
            
        1) Old furniture
        2) Old posters and paintings 
        3) Lamps again. Maybe Bunny and Ducky will behave now. \n"""
            )
            arrived(i=2) #call same function

    #if any other choice (non existent) try again
    else:
        print("""That is not acceptable answer, please try again \n""")
        arrived()


# STAGE 2
# function for when you have to save forky
def inside():
    """Second stage, inside the shop. Get help from one of your friends"""
    print("""\n --- CHECKPOINT --- """)
    print(f"""You walk over to the cabinet and attempt to distract the dummy. Who \nwill you ask for help? (Bunny, Ducky, Bo Peep, Buzz Lightyear, No one) \n*You cannot help two people, it will choose whoever appears first in this list \n""")
    #asking who they want help from (only one choice fails) 
    helper = input(prompt=f"""{">"*80} """)
    option1 = ["bunny", "rabbit", "bun", "buns", "one"]
    option2 = ["ducky", "duck", "quack", "two"]
    option3 = ["bo", "peep", "bopeep", "girlfriend", "three"]
    option4 = ["buzz", "bus", "lightyear", "light", "year", "four"]
    option5 = ["no", "one", "noone", "alone", "five"]

    optionone = any(word in helper.casefold() for word in option1)
    optiontwo = any(word in helper.casefold() for word in option2)
    optionthree = any(word in helper.casefold() for word in option3)
    optionfour = any(word in helper.casefold() for word in option4)
    optionfive = any(word in helper.casefold() for word in option5)

    #if option 1
    if optionone == True or helper == "1":
        print("""He tells you he will only work with Ducky. They both run off and suddenly you hear a big \ncrash. Benson looks over, forgets must to guard the cabinet and rushes off to investigate \nthe noise. He really is a dummy. \n""")
        image("https://i.ytimg.com/vi/w7bTgyJ2yLs/maxresdefault.jpg")
        forky()
    #if option 2
    elif optiontwo == True or helper == "2":
        print("""He tells you he will only work with Ducky. They both run off and suddenly you hear a big \ncrash. Benson looks over, forgets must to guard the cabinet and rushes off to investigate \nthe noise. He really is a dummy. \n""")
        image("https://i.ytimg.com/vi/w7bTgyJ2yLs/maxresdefault.jpg")
        forky()
    #if option 3
    elif optionthree == True or helper == "3":
        print("""Bo grabs a small bead from the ground and uses her cane to fling it at Benson. It smacks him \nright in the face and rolls away. He follows the bead into a narrow space between some books, and you \nlock him inside by pushing a giant stack in front of the exit. \n""")
        image("https://cdn.abcotvs.com/dip/images/5352460_061819-wls-toy-story-bo-peep-430p-vid.jpg")
        forky()
    #if option 4
    elif optionfour == True or helper == "4":
        print("""He thinks for a second and then runs off and grabs a small decoration. He throws it at \nBenson, and everyone ducks as he turns. Benson is very confused and runs off trying to find where \nthe attack came from. You all enter the cabinet. \n""")
        image("https://imagesvc.meredithcorp.io/v3/mm/image?url=https%3A%2F%2Fstatic.onecms.io%2Fwp-content%2Fuploads%2Fsites%2F6%2F2019%2F05%2Ftoystory45bead0596493d-2000.jpg")
        forky()
    #if option 5
    elif optionfive == True or helper == "5":
        print("""You decide to take matters into your own hands and walk up to fight Benson. He is much \nlarger than you. He captures you and calls Gabby Gabby. There is no way to escape. \n""")
        image("https://www.rotoscopers.com/wp-content/uploads/2019/10/Toy-Story-4-Woody--750x375.jpg")
        fail()
    # anything else ask again
    else:
        #if any other choice (non existent) try again
        print(f"""That is not an acceptable answer. Please try again""")
        return inside()


# STAGE 3
# function where you find forky and fight Gabby Gabby
def forky():
    """You find forky and have to get him out of the cabinet he is in"""
    print("""\n --- CHECKPOINT --- """)
    print("""You enter the cabinet and find Forky laying on a bed. He stands up quickly when he sees \nyou and immediately goes in for a hug. \n""")
    image("https://i.ytimg.com/vi/Zz9kyQpfm-I/maxresdefault.jpg")
    #ask if he will hurry bc Gabby Gabby is coming
    decision = input(prompt = f"""Do you tell him you need to hurry up and leave before Gabby Gabby comes? \n{">"*80} """)
    yes = ["yes", "leave", "hurry", "yeah", "yep", "of course"]
    no = ["no", "nope", "nah", "dont", "don't", "negative"]
    hurry = any(word in decision.casefold() for word in yes) # https://www.kite.com/python/answers/how-to-check-if-a-list-contains-a-substring-in-python
    nohurry = any(word in decision.casefold() for word in no)
    #if yes nested loop starts
    if hurry == True or decision == "y": 
        print("You push him off gently and tell him you must hurry and leave before Gabby Gabby comes \nback. You pick him up and start climbing down the cabinet. ")
        #ask if he will throw Forky down
        throw = input(prompt = f"""Do you push forky down the cabinet to leave quickly? (He won't die) \n{">"*80} """)
        yes = ["yes", "quickly", "push", "yeah", "yep", "of course"]
        throws = any(word in throw.casefold() for word in yes) # https://www.kite.com/python/answers/how-to-check-if-a-list-contains-a-substring-in-python
        nothrows = any(word in throw.casefold() for word in no)
        #if yes nested loop continues
        if throws == True or throw == "y": 
            print("You throw Forky down the cabinet and he lands safely... on top of a sleeping cat.")
            jump = input(prompt = f"""Do you jump too? (You won't die) \n{">"*80} """)
            yes = ["yes", "jump", "too", "yeah", "yep", "of course"]
            jumps = any(word in jump.casefold() for word in yes) # https://www.kite.com/python/answers/how-to-check-if-a-list-contains-a-substring-in-python
            nojumps = any(word in jump.casefold() for word in no)
            #last nested if for jumping down
            if jumps == True or jump == "y": 
                print("You jump after Forky, landing on the cat. So does everyone else. It wakes up just as Buzz \nis jumping off the cabinet, but he uses his wings to angle himself onto the cat. You hold on to dear life as \nthe cat runs around the store, knocking antiques and dummies everywhere. When it gets close enough to the exit \nyou jump off, and manage to leave the store unscathed.\n")
                win()
            elif nojumps  == True or jump == "n": 
                print("You slowly climb down the cabinet, quiet enough so the cat does not wake up. \nYou reconvene with all your friends and start walking towards the exit.")
                print("Gabby Gabby stops you in your tracks and starts pulling Forky towards her. Your friends \nhelp pull Forky and when Gabby Gabby lets go she hits a mirror, which falls lightly on her. The mirror stars \nsliding further down. You look at Gabby Gabby's face of despair.")
                escape()
            #anything else replay stage
            else: 
                print(f"""That is not an acceptable answer. Please try again.""")
                return forky()
        elif nothrows == True or throw == "n": 
            print("You start climbing down while holding Forky, but you slip and both of you fall. You land on \nsoft fur. You beckon for your friends to come down too, but when Bunny and Ducky come down they drop a magnet \nand it wakes up the sleeping cat. All your friends jump to sit on it with you, and as it tries to get you off \nhis back he passes the exit and you manage to leave the store. \n")
            win()
        #anything else replay stage
        else: 
            print(f"""That is not an acceptable answer. Please try again.""")
            return forky()
    elif nohurry == True or decision == "n": 
        print(f"""You hug Forky back and take time to tell him about what an adventure it was to come rescue \nhim. When you finally decide to leave Gabby Gabby stops you in your tracks and starts pulling Forky towards \nher. Your friends help pull Forky and when Gabby Gabby lets go she hits a mirror, which falls lightly on her. \nThe mirror starts sliding further down. You look at Gabby Gabby's face of despair.""")
        escape()
    #anything else replay stage
    else: 
        print(f"""That is not an acceptable answer. Please try again.""")
        return forky()
    
    
# STAGE 4
# function for maze bc you took long route
def maze():
    """A maze that makes you get stuck 4 times if you give acceptable answers every time, otherwise it restarts."""
    print("""\n --- CHECKPOINT --- """)
    #options to print during the maze. They may print more than once
    lost = 0
    #while loop for the maze (will run at least three times no matter what)
    while lost < 3: #will get stuck in a maze for 3 rounds regardless of choices
        op1 = f"""You walk {random.randrange(start = 5, stop = 10)} steps. You reach another crossroad. A big blue book looms over you."""
        op2 = f"""You walk {random.randrange(start = 15, stop = 20)} steps. You walk for what seems hours and end up at an intersection. """
        op3 = f"""You walk {random.randrange(start = 5, stop = 15)} steps. You seem to be walking in circles, but you don't remember the name of the book at \nthe entrance of the maze"""
        op4 = f"""You walk {random.randrange(start = 5, stop = 15)} steps. You reach a hallway where all the book covers are green. It reminds you of the grass \nin Andy's house"""
        op5 = f"""You walk {random.randrange(start = 5, stop = 15)} steps. You see light in front and think you are reaching an exit, but it turns out it was a \nsmall opening between two of the old tattered books"""
        op6 = f"""You walk {random.randrange(start = 5, stop = 15)} steps. A book falls behind you, blocking the way back. No choice but to keep going. """

        options = [op1, op2, op3, op4, op5, op6]
        lost = lost + 1
        mazze = input(prompt=f"""Do you turn left, right, or continue straight? \n{">"*80} """)
        option1 = ["one", "1st", "left"]
        option2 = ["two", "second", "right"]
        option3 = ["three", "3rd", "straight"]

        optionone = any(word in mazze.casefold() for word in option1)
        optiontwo = any(word in mazze.casefold() for word in option2)
        optionthree = any(word in mazze.casefold() for word in option3)
        
        #have to give a correct answer for it to work 
        if optionone == True or optiontwo == True or optionthree == True or mazze == "1" or mazze == "2" or mazze == "3"or mazze == "l" or mazze == "r" or mazze == "s": # will get stuck in a maze for 3 rounds regardless of choices - it might repeat the prints
            print(random.choice(seq = options))
        else:
            print("""That is not acceptable answer, please try again from the last checkpoint""")
            maze() #if you make a mistake the entire maze function will start again and you get lost at least 3 more times
    #exit the maze while loop and continue the story (still in the maze but will leave as long as the answer is accepted)
    first = input(prompt=f"""Do you turn left, right, or continue straight? \n{">"*80} \n""")
    optionone = any(word in first.casefold() for word in option1)
    optiontwo = any(word in first.casefold() for word in option2)
    optionthree = any(word in first.casefold() for word in option3)
    if optionone == True or optiontwo == True or optionthree == True or first == "1" or first == "2" or first == "3"or first == "l" or first == "r" or first == "s":
        print(f"""\nYou walk {random.randrange(start = 1, stop = 10)} steps. You finally reach an exit. You are at the bottom of a large cabinet. You see Gabby Gabby and \nBenson walking above. They have not seen you. You knock over some of the maze books and begin climbing the \ncabinet while Gabby Gabby and Benson check the maze. \n""")
        forky()
    #if unacceptable answer start from last checkpoint
    else:
        print("""That is not acceptable answer, please try again from the last checkpoint""")
        maze()


# STAGE 5
# function where you escape?
def escape():
    """A last decision of wether to help Gabby Gabby or not"""
    print("""\n --- CHECKPOINT --- """)
    #ask if he will help. Double p because help is a function and it fails
    helpp = input(prompt = f"""Will you help Gabby Gabby? \n{">"*80} \n""")
    option1 = ["yes", "course", "yep", "definitely", "ofcourse", "yup", "yeah"]
    option2 = ["no", "nope", "nah", "dont", "don't", "negative"]
    helps = any(word in helpp.casefold() for word in option1)
    nohelps = any(word in helpp.casefold() for word in option2)
    #if you help you win but save Gabby Gabby
    if helps == True or helpp == "y":
        print(f"""Hurry! You must guess a number between one and ten correctly to pull with the necessary strength to lift the mirror. """)
        answer = str(random.randint(a = 1, b = 10)) 
        guess = input(prompt = f"""{">" * 80} \n""")
        while guess != answer:
            guess = input(prompt = f"""\nSorry, you will have to keep pulling. How much strength should you apply? \n{">" * 80} \n""")
        image("https://i.ytimg.com/vi/E-3PihCoI0Y/hqdefault.jpg")
        print("""You pull Gabby Gabby out and the mirror falls where she laid and shatters. She thanks you, \nand you offer to take her home with you. She accepts, and guides you towards the door. \n""")
        win()
    #if you dont help you still win but you feel bad that she dies
    elif nohelps == True or helpp == "n":
        print("""The mirror falls on top of Gabby Gabby as you walk away. You never hear from her again.""")
        win()
    #unaccpetable answers get asked again
    else:
        print("That is not an acceptable answer. Try again \n")
        escape()


        
        
print(game.__doc__)
game()
```
